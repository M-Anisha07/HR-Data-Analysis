```python
#Attrition count
```

```python
attrition_counts = df['attrition'].value_counts()

plt.figure()
plt.bar(attrition_counts.index, attrition_counts.values)
plt.title("Employee Attrition Count")
plt.xlabel("Attrition")
plt.ylabel("Number of Employees")
plt.show()
```

```python
#Attrition by Age Group
```

```python
bins = [18, 25, 35, 45, 55, 65]
labels = ['18-25', '26-35', '36-45', '46-55', '55+']

df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)
age_attrition = df[df['attrition'] == 'Yes']['age_group'].value_counts().sort_index()

plt.figure()
plt.bar(age_attrition.index, age_attrition.values)
plt.title("Attrition by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Employees Left")
plt.show()
```

```python
#Attrition Rate
```

```python
attrition_counts = df['attrition'].value_counts()

plt.figure()
plt.pie(
    attrition_counts,
    labels=attrition_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Employee Attrition Distribution")
plt.show()
```

```python
#Salary vs Attrition
```

```python
import seaborn as sns
plt.figure()
sns.boxplot(x='attrition', y='monthly_income', data=df)
plt.title("Monthly Income vs Attrition")
plt.show()
```

```python
#Female vs Male Attrition

attrition_df = df[df['attrition'] == 'Yes']
gender_counts = attrition_df['gender'].value_counts()

plt.figure()

plt.pie(
    gender_counts.values,
    labels=gender_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops={'width': 0.4} )

plt.title("Female vs Male Attrition")
plt.show()
```
